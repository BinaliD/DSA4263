```python
import pandas as pd
import numpy as np
import requests
from io import StringIO



```

```python
# Load genuine user data and bot data from GitHub
genuine_url = "https://raw.githubusercontent.com/BinaliD/DSA4263/main/GenuineUsers.csv"
bot_url = "https://raw.githubusercontent.com/BinaliD/DSA4263/main/SocialSpamBotUsers.csv"
```

```python
genuine_data = pd.read_csv(genuine_url)
bot_data = pd.read_csv(bot_url)
```

```python
# Randomly sample 5000 rows from each dataset
genuine_sample = genuine_data.sample(n=500, random_state=42)
bot_sample = bot_data.sample(n=500, random_state=42)

# Add a new column to each sample to indicate bot or genuine
genuine_sample['Type'] = 'Genuine'
bot_sample['Type'] = 'Bot'

# Concatenate the two samples into a single dataframe
combined_data = pd.concat([genuine_sample, bot_sample])

# Shuffle the combined dataframe
combined_data = combined_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the combined dataframe to a new CSV file
combined_data.to_csv("combined_data.csv", index=False)
```
